# **Regla de tres simple**

## **Ejemplos**

#### **Directa**

### **Ejemplo 1**

```python
# Si 12 discos compactos cuestan $600, ¿cuánto costarán 18?
num_discos = 12
precio = 600
num_discos2 = 18

x = (num_discos2 * precio) / num_discos
print(x)
```

### **Ejemplo 2**

```python
# Una llave que se abre 4 horas diarias durante 5 días, vierte 5 200 litros de agua, ¿cuántos litros vertirá en 12 días si se abre 4 horas por día?
# 

x = 4 * 5
y = 5200
z = 4 * 12
print(f"{x} horas, {y} litros de agua, {z} horas")
print(f"{x}/{z}, {y}/?")

incog = (z * y)/x
print(f"{int(incog)} litros de agua")
```

#### **Inversa**

```python
# Se ha planeado que una barda sea construida por 24 hombres en 18 días; sin embargo, sólo se logró contratar a 12 hombres, ¿en cuántos días la construirán?

# 24/12, 18/? -> ?/18

x = 24 # 24 hombres
y = 18 # 18 días
z = 12 # 12 hombres

incog = (y * x)/z
print(f"{int(incog)} días")
```

```python
# Las ruedas traseras y delanteras de un automóvil tienen un diámetro de 1.5 m y 1 m, respectivamente, cuando las primeras han dado 350 vueltas, ¿cuántas han dado las segundas?

# 1.5/1, 350/? -> ?/350
x, y , z = 1.5, 1, 350
incog = (z * x)/y
print(f"{int(incog)} vueltas" )
```

## **Ejercicio 81**

```python
# EJERCICIO 81
# Resuelve los siguientes problemas:

# 1. El precio de 25 latas de aceite es de $248,
#    ¿cuántas latas se podrán comprar con $1240?

x, y = 248, 1240
z = 25

incog = (z * y)/x
print(f"{int(incog)} latas")
```

```python
# 2. Liam escucha la radio durante 30 minutos, lapso en el que hay 7 minutos
#    de anuncios comerciales; si escucha la radio durante 120 minutos,
#    ¿cuántos minutos de anuncios escuchará?

x, y = 30, 120
z = 7
incog = (z * y)/x
print(f"{int(incog)} minutos de anuncios")
```

```python
# 3. Durante 70 días de trabajo Ana ganó $3500,
#    ¿cuánto ganaría si trabajara 12 días más?

x, y = 70, (70+12)
z = 3500
incog = (z * y)/x
print(f"Ganaría ${int(incog)}")
```

```python
# 4. Una llave abierta 6 horas diarias durante 7 días arrojó 6120 litros de agua,
#    ¿cuántos litros arrojará durante 14 días si se abre 4 horas diarias?

x, y = (6*7), (4*14)
print(f"6 horas diaras durante 7 días son: {x} horas\n4 horas diarias durante 14 días son: {y} horas")
z = 6120
incog = (z * y)/x
print(f"{int(incog)} litros de agua")
```

```python
# 5. Un automóvil gasta 9 litros de gasolina cada 120 km.
#    Si quedan en el depósito 6 litros, ¿cuántos kilómetros podrá recorrer?

x, y = 9, 6
z = 120
incog = (z * y)/x
print(f"{int(incog)} km podrá recorrer")
```

```python
# 6. En un libro de 80 páginas cada una tiene 35 líneas,
#    ¿cuántas páginas tendrá el mismo libro si en cada una se colocan 40 líneas?

x, y = 35, 40
z = 80
incog = (z * x)/y
print(f"{int(incog)} páginas")
```

```python
# 7. Una bodega se llena con 3500 sacos de 6 kg de papas cada uno y otra de la misma capacidad
#    se llena con sacos de 5 kg, ¿cuántos sacos caben en la segunda bodega?

x, y = 6, 5
z = 3500
incog = (z * x) / y
print(f"{int(incog)} sacos en la bodega")
```

```python
# 8. Un leñador tarda 8 segundos en dividir en 4 partes un tronco de cierto tamaño,
#    ¿cuánto tiempo tardará en dividir un tronco semejante en 5 partes?

x, y = 4, 5
z = 8
incog = (z * y)/x
print(f"{incog} segundos")
```

```python
# 9. Si un automóvil hizo 9 horas durante un recorrido de 750 km,
#    ¿qué tiempo empleará en recorrer 2250 km si su velocidad es constante?

x, y = 750, 2250
z = 9
incog = (z * y)/x
print(f"{incog} horas")
```
