## **Ejercicio 78**

```python
# EJERCICIO 78
# Encuentra la media proporcional (geométrica) entre los números dados:

# 1. 12 y 3
x = sp.sqrt(12)
y = sp.sqrt(3)
med_geom = x * y
print(f"Media geométrica de 12 y 3 es: {med_geom}")

# 2. 6 y 24
x = sp.sqrt(6)
y = sp.sqrt(24)
med_geom = x * y
print(f"La media geométrica entre 6 y 24 es: {med_geom}")

# 3. 9 y 25
x = sp.sqrt(9)
y = sp.sqrt(25)
med_geom = x * y
print(f"La media geométrica entre 9 y 25 es: {med_geom}")

# 4. 4 y 12
x  = sp.sqrt(4)
y = sp.sqrt(12)
med_geom = x * y
print(f"La media geométrica entre 4 y 12 es: {med_geom}")

# 5. 2 y 7
x = sp.sqrt(2)
y = sp.sqrt(7)
med_geom = x * y
print(f"La media geométrica entre 2 y 7 es: {med_geom}")
```

```python
# 6. 9 y 18
x, y =  sp.sqrt(9), sp.sqrt(18)
med_geom = x * y
print(f"La media geométrica entre 9 y 18 es: {med_geom}")

# 7. 10 y 25
x, y = sp.sqrt(10), sp.sqrt(25)
med_geom = x * y
print(f"La media geométrica entre 10 y 25 es: {med_geom}")

# 8. 0.1 y 0.5
x = Fraction(1, 10)
y = Fraction(1, 2)

med_geom = sp.sqrt(x * y)

print(f"La media geométrica entre 0.1 y 0.5 es: {med_geom}")
print(float(med_geom))  # valor decimal

# 9. 0.2 y 0.8
x = 0.2
y = 0.8
med_geom = sp.sqrt(x * y)
print(f"La media geométrica entre 0.2 y 0.8 es: {float(med_geom):.2f}")

# 10. 0.8 y 1.6
x = 0.2
y = 0.8
med_geom = sp.sqrt(0.8, 1.6)
print(f"La media geométrica entre 0.8 y 1.6 es: {float(med_geom):.2f}")
```
