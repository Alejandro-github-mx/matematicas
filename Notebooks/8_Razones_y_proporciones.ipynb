### **Ejemplo 2**

```python
x, y, z = 5, 24, 30
m = (x * y) / z
print(f"El valor de m es: {int(m)}")
```

### **Ejemplo 3**

```python
x, y, z = 10, 2, 7
b = (x * y)/z
print(f"El valor de b es: {float(b):.2f}")
fraction = Fraction(b).limit_denominator()
print(f"El valor en fracción de b es: {fraction}")
```
