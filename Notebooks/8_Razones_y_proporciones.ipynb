### **Ejemplo 2**

```python
# Una llave que se abre 4 horas diarias durante 5 días, vierte 5 200 litros de agua, ¿cuántos litros vertirá en 12 días si se abre 4 horas por día?
# 

x = 4 * 5
y = 5200
z = 4 * 12
print(f"{x} horas, {y} litros de agua, {z} horas")
print(f"{x}/{z}, {y}/?")

incog = (z * y)/x
print(f"{int(incog)} litros de agua")
```

#### **Inversa**

```python
# Se ha planeado que una barda sea construida por 24 hombres en 18 días; sin embargo, sólo se logró contratar a 12 hombres, ¿en cuántos días la construirán?

# 24/12, 18/? -> ?/18

x = 24 # 24 hombres
y = 18 # 18 días
z = 12 # 12 hombres

incog = (y * x)/z
print(f"{int(incog)} días")
```

```python
# Las ruedas traseras y delanteras de un automóvil tienen un diámetro de 1.5 m y 1 m, respectivamente, cuando las primeras han dado 350 vueltas, ¿cuántas han dado las segundas?

# 1.5/1, 350/? -> ?/350
x, y , z = 1.5, 1, 350
incog = (z * x)/y
print(f"{int(incog)} vueltas" )
```

# **Regla de tres compuesta**

## **Ejemplos**

### **Ejemplo 1**

```python
# Una guardería con 250 niños proporciona 4 raciones de alimentos diarios a cada pequeño durante 18 días. Si la población aumenta a 50 niños, ¿cuántos días durarán los alimentos si se disminuyen a 3 raciones diarias?
niños1, niños2 = 250, 300
raciones1, raciones2 = 4, 3
dias1 = 18
dias2 = (niños1 * raciones1 * dias1) / (raciones2 * niños2)
print(f"{dias2}")
```

### **Ejemplo 2**

```python
# 15 cajas de aceite con 18 galones cuestan $960, ¿cuánto cuestan 9 cajas con 20 galones?
caja1, caja2 = 15, 9
content1, content2 = 18, 20
precio1 = 960

precio2 = (caja2 * content2 * precio1) / (caja1 * content1)
print(f"9 cajas cuestan ${precio2}")
```

### **Ejemplo 3**

```python
# Se calcula que para construir una barda de 600 m en 18 días, trabajando 8 horas diarias, se necesitan 12 hombres, ¿cuántos días tardarán 8 hombres trabajando 6 horas diarias para construir una barda de 400 m?

barda1, barda2 = 600, 400
hombres1, hombres2 = 12, 8
horas1, horas2 = 8, 6
dias1 = 18

dias2 = dias1 * (hombres1 / hombres2) * (horas1 / horas2) * (barda2 / barda1)
print(f"8 hombres se tardarán {int(dias2)} dias trabajando {horas2} diarias")
```

# **Interés, capital y tiempo**

### **Funciones para el cálculo de interés**

```python
# Si el tiempo está en años
def interes_simple_años(capital, tasa, tiempo):
    return (capital * tasa * tiempo)/100

# si el tiempo está en meses
def interes_simple_meses(capital, tasa, meses):
    return (capital * tasa * meses)/1200

# si el tiempo está en días
def interes_simple_dias(capital, tasa, dias):
    return (capital * tasa * dias)/36000
```

### **Funciones para el cálculo del capital, tiempo y tasa**

```python
# Capital
def capital_en_años(interes, tiempo, tasa):
    return (100*interes)/(tiempo * tasa)

def capital_en_meses(interes, tiempo, tasa):
    return (1200*interes)/(tiempo*tasa)

def capital_en_dias(interes, tiempo, tasa):
    return(36000*interes)/(tiempo*tasa)

# Tiempo
def tiempo_en_años(interes, capital, tasa):
    return(100*interes)/(capital*tasa)

def tiempo_en_meses(interes, capital, tasa):
    return(1200*interes)/(capital*tasa)

def tiempo_en_dias(interes, capital, tasa):
    return(36000*interes)/(capital * tasa)

# Tasa
def tasa_en_años(interes, capital, tiempo):
    return (100*interes)/(capital * tiempo)

def tasa_en_meses(interes, capital, tiempo):
    return (1200*interes)/(capital * tiempo)

def tasa_en_dias(interes, capital, tiempo):
    return (36000*interes)/(capital * tiempo)
```
