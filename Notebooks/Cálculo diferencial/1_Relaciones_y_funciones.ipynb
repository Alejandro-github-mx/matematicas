#### **Función para ver si es función o no**

```python
def es_funcion1(relacion):
    entradas = {}
    
    for x, y in relacion:
        if x in entradas and entradas[x] != y:
            return False
        entradas[x] = y
    return True
```

```python
def es_funcion2(relacion):
    entradas = {}
    
    for x, y in relacion:
        print("Estoy revisando el par:", (x, y))
        print("Entradas actuales:", entradas)
        
        if x in entradas:
            print(x, "ya estaba guardado")
            
            if entradas[x] != y:
                print("Pero antes tenía", entradas[x], "y ahora tiene", y)
                return False
        
        entradas[x] = y
        print("Ahora guardo:", x, "→", y)
        print("-----")
    
    return True
```

### **Ejemplo 1**

```python
A = [(2, 4), (3, 9), (4, 16), (5, 25)]
B = [(3, 2), (3, 6), (5, 7), (5, 8)]
C = [(2, 4), (3, 4), (5, 4), (6, 4)]
M = [(2, 4), (6, 2), (7, 3), (4, 12), (2, 6)]
print("A:", es_funcion1(A))
print("B:", es_funcion1(B))
print("C:", es_funcion1(C))
print("M:", es_funcion1(M))
```
