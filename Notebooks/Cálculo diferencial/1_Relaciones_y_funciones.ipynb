#### **Función para ver si es función o no**

```python
def es_funcion1(relacion):
    entradas = {}
    
    for x, y in relacion:
        if x in entradas and entradas[x] != y:
            return False
        entradas[x] = y
    return True
```

```python
def es_funcion2(relacion):
    entradas = {}
    
    for x, y in relacion:
        print("Estoy revisando el par:", (x, y))
        print("Entradas actuales:", entradas)
        
        if x in entradas:
            print(x, "ya estaba guardado")
            
            if entradas[x] != y:
                print("Pero antes tenía", entradas[x], "y ahora tiene", y)
                return False
        
        entradas[x] = y
        print("Ahora guardo:", x, "→", y)
        print("-----")
    
    return True
```

### **Función para ver si es relación o función 3**

```python
def es_funcion3(relacion):
    entradas = {}
    
    for x,y in relacion:
        x_simplificado = sp.simplify(x) 
        y_simplificado = sp.simplify(y)
    
        if x_simplificado in entradas and entradas[x_simplificado] != y_simplificado:
            return False
        entradas[x_simplificado] = y_simplificado
    
    return True       
```

# **Valor de una función**

## **Ejemplos**

### **Ejemplo 1**

```python
x, a, b, t = sp.symbols('x a b t')
```

```python
# 1. Obtén f(-3) para f(x) = 3x^2 - 5x - 2

f1 = 3*x**2 - 5*x -2
resultado1 = f1.subs(x, -3)

print(f"1) f(-3): {resultado1}")
```

```python
# 2. Si f(x) = (3x - 1) / (5 - x), encuentra f(3/4)

f2 = (3*x - 1) / (5 - x)
resultado2 = sp.simplify(f2.subs(x, sp.Rational(3,4)))
print(f"2) f(3/4): {resultado2}")
```

```python
# 3. Si s(t) = sqrt(t - 5), determina s(4), s(a + 5)

s = sp.sqrt(t - 5)
resultadoa = s.subs(t, 4)
resultadob = sp.simplify(s.subs(t, a +5))
print("3) s(4) =", resultadoa)
print("   s(a + 5) =", resultadob)
```

```python
# 4. Si f(x) = sen(x + pi/4), determina f(pi/3)

f4 = sp.sin(x + sp.pi/4)
resultado4 = sp.simplify(f4.subs(x, sp.pi/3))
print(f"4) f(pi/3): {resultado4}")
```

```python
# 5. Determina (f(a + b) - f(a)) / b si f(x) = sqrt(x)

f5 = sp.sqrt(x)
resultado5 = sp.simplify((f5.subs(x, a + b) - f5.subs(x, a)) / b)
print("5) (f(a+b)-f(a))/b =", resultado5)
```

```python
# 6. Si y = x / (x + 2), encuentra el valor de y cuando x = -2

y = x / (x + 2)
resultado6 = y.subs(x, -2)
print("6) y cuando x = -2 =", resultado6)
```

```python
# 7. Si f(x) = x^2 - 1, demuestra que f(1/x) = -f(x) / x^2

f7 = x**2 - 1
lado_izq = sp.simplify(f7.subs(x, 1/x))
lado_der = sp.simplify(-f7 / x**2)
print("7) f(1/x) =", lado_izq)
print("   -f(x)/x^2 =", lado_der)
print("   ¿Son iguales?", sp.simplify(lado_izq - lado_der) == 0)
```

# **Dominio, contradominio y rango de una función**

## **Ejemplos**

```python
# 1. ¿Cuál es el dominio de la función f(x) = 3x^2 - 5x - 6?
x = sp.symbols('x')

f1 = 3*x**2 - 5*x - 6
print(sp.calculus.util.continuous_domain(f1, x, sp.S.Reals))
```

```python
# 2. Determina el dominio de la función f(x) = (x - 3) / (x + 5)

f2 = (x-3)/(x+5)
print(sp.calculus.util.continuous_domain(f2, x, sp.S.Reals))
```

```python
# 3. ¿Cuál es el dominio de la función f(x) = x / (x^2 - 5x - 6)?

f3 = x /(x**2 - 5*x - 6)
print(sp.factor(x**2 - 5*x - 6))
print(sp.calculus.util.continuous_domain(f3, x, sp.S.Reals))
```
