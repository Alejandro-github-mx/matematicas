### **Función para ver si es función o no**

```python
def es_funcion1(relacion):
    entradas = {}
    
    for x, y in relacion:
        if x in entradas and entradas[x] != y:
            return False
        entradas[x] = y
    return True
```

```python
def es_funcion2(relacion):
    entradas = {}
    
    for x, y in relacion:
        print("Estoy revisando el par:", (x, y))
        print("Entradas actuales:", entradas)
        
        if x in entradas:
            print(x, "ya estaba guardado")
            
            if entradas[x] != y:
                print("Pero antes tenía", entradas[x], "y ahora tiene", y)
                return False
        
        entradas[x] = y
        print("Ahora guardo:", x, "→", y)
        print("-----")
    
    return True
```
