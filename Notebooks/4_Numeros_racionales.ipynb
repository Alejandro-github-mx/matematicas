## Ejercicio de Práctica 1

```python
# Fracción 3/4
# La fracción $\frac{3}{4}$ significa que la unidad se divide en 4 partes iguales y se toman 3.

numerador = 3
denominador = 4

print(f"La fracción es {numerador}/{denominador}")
```

```python
fig, ax = plt.subplots(figsize=(8,2))

# Dibujar los 4 cuartos

for i in range(denominador):
  if i < numerador:
    color="lightgray"
  else:
    color= "white"
  
  rect = Rectangle((i, 0), 1, 1, edgecolor="black", facecolor=color)
  ax.add_patch(rect)
  ax.text(i + 0.05, 0.05, "1/4", ha = "center", va = "center", fontsize = 14)
  
ax.set_xlim(0, denominador)
ax.set_ylim(0,1)
ax.set_aspect("equal")
ax.axis("off")

plt.title("Representación gráfica de 3/4")
plt.show()
```

```python
numerador = 3
denominador = 4

valores = [1] * denominador
colores = ["lightgray"] * numerador + ["white"] * (denominador - numerador)

plt.figure(figsize=(8, 2))
plt.bar(range(denominador), valores, width=1, edgecolor="black", color=colores)

for i in range(denominador):
    plt.text(i, 0.5, "1/4", ha="center", va="center", fontsize=14)

plt.xlim(-0.5, denominador - 0.5)
plt.ylim(0, 1)
plt.axis("off")
plt.title("Representación gráfica de 3/4")
plt.show()
```

## **Fracciones equivalentes**

### **Función para fracciones equivalentes**

```python
def fracc_equiva(numerador1, denominador1, numerador2, denominador2):
  resultado1 = numerador1 * denominador2
  resultado2 = denominador1 * numerador2
  if resultado1 == resultado2:
    print(f"Las fracciones son equivalentes: {resultado1} == {resultado2}")
  else:
    print("Las fracciones no son equivalentes") 
```

```python
# 1) 2/5 y 6/15
fracc_equiva(2,5,6,15)
# 2) 3/8 y 48/17
fracc_equiva(3,8,48,17)
# 3) 1/6 y 12/72
fracc_equiva(1,6,12,72)
# 4) 4/9 y 28/72
fracc_equiva(4,9,28,72)
# 5) 18/24 y 6/8
fracc_equiva(18,24,6,8)
```

## **Simplificar fracciones**

```python
# Simplifica la fracción 12/14
def simplificar(numerador,denominador):
  mcd = math.gcd(numerador,denominador)
  return numerador // mcd, denominador // mcd
```

## **Suma y resta con igual denominador**

### **Función suma y resta con igual denominador**

```python
def suma_fracc(numeradores, denominador):
  suma = sum(numeradores)
  
  mcd= math.gcd(suma, denominador)
  n = suma // mcd
  d = denominador // mcd
  
  print(f"Suma: {suma}/{denominador}")
  print(f"Resultado simplificado: {n}/{d}")
```

### **Ejemplo 1**

```python
# Efectúa la operación 3/4 + 2/4 + 1/4

suma_fracc([3,2,4], 4)
```

```python
# EJERCICIO 35
# Resuelve los siguientes problemas:

# 1) Juan compró en el supermercado 1/2 kg de azúcar, 3/4 kg de harina y 1 kg de huevo,
#    estos productos los colocó en una bolsa, ¿cuántos kilogramos pesó dicha bolsa?
suma = sum([1,3,1])
mcd= math.gcd(suma, 4)
n = suma // mcd
d = 4 // mcd

print(f"Suma: {suma}/{4}")
print(f"Resultado simplificado: {n}/{d}")
```

```python
# 2) Dos calles tienen las siguientes longitudes: 2 2/5 y 1 3/4 de kilómetro,
#    ¿cuál es la longitud total de ambas?

## Fracción uno
def convers_mixtas(entero, numerador, denominador):
  return (entero * denominador) + numerador, denominador

n1,d1 = convers_mixtas(2,2,5)
print(f"Fracción uno: {n1}/{d1}")
n2,d2 = convers_mixtas(1,3,4)
print(f"Fracción dos: {n2}/{d2}")

# Denominador común
den_comun = math.lcm(d1, d2)
# Convertir a mismo denominador
nuevo_n1 = n1 * d2
nuevo_n2 = n2 * d1
# Suma numeradores
suma_numeradores = nuevo_n1 + nuevo_n2
mcd = math.gcd(suma_numeradores, den_comun)
n = suma_numeradores // mcd
d = den_comun // mcd
print(f"Suma: {n}/{d}")

# Convertir a mixta
entero = n // d
residuo = n % d

print(f"Resultado final: {entero} {residuo}/{d} km")




```

```python
# 2) Dos calles tienen las siguientes longitudes: 2 2/5 y 1 3/4 de kilómetro,
#    ¿cuál es la longitud total de ambas?

fracc1 = 2 + Fraction(2,5)
fracc2 = 1 + Fraction(3,4)

total = fracc1 + fracc2

print(f"Longitud total de ambas: {total} de km" )
print(f"Como número mixto: {total.numerator // total.denominator} {total.numerator % total.denominator}/{total.denominator} km")
```

```python
# 4) A Joel le pidieron que realizara una tarea de física que consistía en contestar un cuestionario
#    y resolver unos problemas. Se tardó 3/4 de hora en responder el cuestionario y 2 1/2 para
#    solucionar los problemas, ¿cuánto tiempo le tomó a Joel terminar toda la tarea?

cuestio = Fraction(3,4)
probl = 2 + Fraction(1,2)
total = cuestio + probl

print(f"A Joel le tomó {total.numerator // total. denominator} {total.numerator % total.denominator}/{total.denominator} horas")
```

## **Ejercicio 37**

```python
# 1) Una alberca tiene capacidad para 3000 litros de agua, si sólo se encuentra
#    a tres cuartas partes de su capacidad, ¿cuántos litros tiene?

total_alberca = Fraction(300,1)
capacidad = Fraction(3,4)
actual_alberca = total_alberca * capacidad

print(f"La alberca tiene {actual_alberca} litros")
```

```python
# 2) En un estadio de béisbol, 2/3 de los aficionados apoyan al equipo local,
#    si el número de asistentes es de 6300 personas,
#    ¿cuántas apoyan al equipo visitante?

num_asistentes = 6300
equi_local = Fraction(2,3)

equi_visitante = num_asistentes - (equi_local * num_asistentes)

print(f"Apoyan al equipo visitante {equi_visitante} personas")
```

```python
# 3) La tercera parte de una población de 2100 habitantes es afectada por cierto virus,
#    ¿cuántos habitantes no padecen el virus?

poblac = Fraction(2100,1)
poblac_infec = Fraction(1,3)

sanos = 2100 - (poblac * poblac_infec)

print(f"La población no infectada es de: {sanos}")
```

```python
# 4) Se sabe que los viernes por la noche en el D.F. 1/4 del total de automovilistas
#    manejan en estado de ebriedad, si se realiza un sondeo entre 600 conductores
#    un viernes por la noche, ¿cuántos automovilistas se espera que manejen en estado inconveniente?

borrachos = Fraction(1,4)
sondeo = Fraction(600,1)

total_borrachos = sondeo * borrachos

print(f"Total de personas en estado inconveniente esa noche: {total_borrachos}")
```

```python
# 5) En una caja hay 120 pelotas: verdes, rojas y azules,
#    si las pelotas rojas son la tercera parte del total y las azules equivalen
#    a la sexta parte, ¿cuántas hay de cada color?

total_pelotas = Fraction(120,1)
rojas = Fraction(1,3) * total_pelotas
azules = Fraction(1,6) * total_pelotas
verdes = total_pelotas - (rojas + azules)
print(f"Pelotas rojas: {rojas}\n Pelotas azules: {azules}\n Pelotas verdes: {verdes}")
```

```python
# 3) ¿Cuántas piezas de 2 2/3 de metro de longitud se obtienen de una varilla
#    de 13 1/3 metros de largo?

pieza = 2 + Fraction(2,3)
varilla = 13 + Fraction(1,3)
pcs = varilla / pieza
print(f"Salen {pcs} piezas")
```
