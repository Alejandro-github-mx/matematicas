## **Divisibilidad de factores primos**

### Realiza la descomposición en sus factores primos de los siguientes números

```python
def factor_paso_a_paso(n):
  if n == 0:
    raise ValueError("0 No tiene factorización prima")
  if n < 0:
    print("n es negativo, saco -1 como factor.")
    print("-1")
    n = -n
  
  factores = []
  
  print("Empiezo con n =", n)
  print("Voy probando divisores primos: 2,3,5,7,11,...")
  
  d = 2
  while d * d <= n:
    while n%d == 0:
      print(str(n) + " ÷ " + str(d) + " = " + str(n // d)) 
      factores.append(d)
      n = n // d
      d = 3 if d == 2 else d + 2  # después del 2, solo impares

    if n > 1:
        print("Factor primo final:", n)
        factores.append(n)

    print("Factores primos (con repetición):", factores)
    return factores
```

```python
print("Descomposición de 72:",factor_paso_a_paso(72))
```

```python
print("Descomposición de 96:", factor_paso_a_paso(96))
```

```python
print("Descomposición de 96:", factor_paso_a_paso(124))
```

```python
def factorizar(n):
  # Caso especial
  if n == 0:
    raise valueError("0 no tiene descomposición en factores primos")
  # Caso negativo
  if n < 0:
    print("n es negativo, saco -1 como factor")
    print("-1")
    n = -n # Convierte a valor absoluto
  
  factores = [] # Aquí se van guardando los factores
  d = 2 #Divisor actual
  
  # Mensajes para seguir el proceso
  print("Empiezo con:", n)
  print("Divisor actual:", d)
  print("-" * 30)
  
  while n > 1: # Mientras n sea mayor que 1, sigue trabajando
    if n % d == 0:
      print(f"{n} / {d} = {n} // {d}")
      factores.append(d)
      n = n // d
    else:
      d+=1
      print("No divide exacto. Cambio divisor a", d)
  
  print("-" * 30)
  print("factores primos:", factores)
  return factores
```

```python
factorizar(100)
```

```python
factorizar(72)
```

```python
from sympy import factorint
nums = [72, 96, 225, 576, 945, 210, 840, 2310, 3675, 2376, 7020, 29400, 30240, 16200, 30030]

def formato(f):
    return " · ".join([f"{p}^{e}" if e > 1 else f"{p}" for p, e in sorted(f.items())])

for n in nums:
    print(f"{n} = {formato(factorint(n))}")
```
