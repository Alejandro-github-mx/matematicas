# **Máximo común divisor (MCD)**

```python
# Es el mayor de los divisores en común de 2 o más números.
```

# **Packages**

```python
import math
```

# **Ejercicios**

```python
#  Encuentra el máximo común divisor de 48, 36 y 60.

math.gcd(48, 36, 60)
```

```python
# Determina el MCD(72,180).

math.gcd(72,180)
```

```python
# Calcula el MCD(11,23).

math.gcd(11, 23)
```

```python
# Encuentra el máximo común divisor de 234, 390 y 546.

math.gcd(234, 390, 546)
```
