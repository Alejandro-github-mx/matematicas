# **Problemas de aplicación**

```python
# En una reunión de academia del área de matemáticas se repartieron 18 bocadillos, 24 vasos con refresco y
# 12 rebanadas de pastel, ¿cuántos profesores asistieron a la reunión y qué cantidad de bocadillos, vasos con refresco y rebanadas de pastel recibió cada uno?

num_profesores = math.gcd(18,24,12)
bocadillos = 18
vasos = 24
rebanadas = 12

print("Número de profesores:", num_profesores)
print("Bocadillos por profesor: ", bocadillos / num_profesores)
print("Vasos de refresco por profesor:", vasos / num_profesores)
print("Número de rebanadas de pastel por profesor:", rebanadas / num_profesores)

```

```python
# Tres escuelas deciden hacer una colecta de dinero entre sus alumnos para donar a varias instituciones de beneficencia. Si la primera junta 120 mil, la segunda 280 mil y la tercera 360 mil pesos, ¿cuál es la mayor cantidad que recibirá cada institución de tal manera que sea la misma y cuántas instituciones podrán ser beneficiadas?

primera_escuela = 120000
segunda_escuela = 280000
tercera_escuela = 360000
max_quant = math.gcd(primera_escuela, segunda_escuela, tercera_escuela)

print("Máx cantidad:", max_quant)

print("Cantidad de escuelas beneficiadas por la primera escuela:", primera_escuela / max_quant)
print("Cantidad de escuelas beneficiadas por la segunda escuela:", segunda_escuela / max_quant)
print("Cantidad de escuelas beneficiadas por la tercera escuela:", tercera_escuela / max_quant)
```

```python
# Al hacer el corte del día en un restaurante, el administrador hace 3 rollos de billetes de la misma denominación, en el primero hay $1 350, en el segundo $1 700 y en el tercero $3 550, ¿cuántos billetes hay en cada rollo y de qué denominación son?

primer_rollo = 1350
segundo_rollo = 1700
tercer_rollo = 3550
todos_billetes = math.gcd(primer_rollo, segundo_rollo, tercer_rollo)

print("¿Cuántos billetes hay en cada rollo?", todos_billetes)
print("Número de billetes en rollo 1:", primer_rollo / todos_billetes)
print("Número de billetes en rollo 2:", segundo_rollo / todos_billetes)
print("Número de billetes en rollo 3:", tercer_rollo / todos_billetes)
```

```python
# Una persona viaja a la Ciudad de México cada 12 días, otra lo hace cada 20 días y una tercera cada 6 días. Si hoy han coincidido en estar las 3 en la ciudad, ¿dentro de cuántos días, como mínimo, volverán a coincidir?

persona_uno = 12
persona_dos = 20
persona_tres = 6
num_dias = math.lcm(persona_uno, persona_dos, persona_tres)

print(f"Las personas van a volver a coincidir en {num_dias} días")
```

```python
# Un médico receta a un paciente tomar una pastilla cada 6 horas y un jarabe cada 8 horas. Si al iniciar el tratamiento toma la pastilla y el jarabe a la misma hora, ¿después de cuántas horas volverá a tomar ambos medicamentos al mismo tiempo?

pastilla = 6
jarabe = 8
num_horas = math.lcm(pastilla, jarabe)

print(f"Después de {num_horas} horas, volverá a tomar los medicamentos al mismo tiempo")
```

```python
# 1) Tres cajas contienen, cada una, 12 kilogramos de carne de res, 18 de carne de cerdo y 24 de carne de pollo. La carne de cada caja está contenida en bolsas del mismo tamaño y con la máxima cantidad de carne posible, ¿cuánto pesa cada bolsa y cuántas hay por caja?

res = 12
cerdo = 18
pollo = 24
cada_bolsa = math.gcd(res, cerdo, pollo)

print(f"Cada bolsa pesa: {cada_bolsa} kilogramos")
print(f"Cantidad de bolsas en caja de res: {res/cada_bolsa}" )
print(f"Cantidad de bolsas en caja de cerdo: {cerdo/cada_bolsa}" )
print(f"Cantidad de bolsas en caja de pollo: {pollo/cada_bolsa}")
```

```python
# 4) Un ciclista da una vuelta a una pista en 6 minutos, mientras que otro tarda 4 minutos. Si ambos inician sus recorridos juntos, ¿después de qué tiempo volverán a encontrarse y cuántas vueltas habrán dado cada uno?

primer_ciclista = 6
segundo_ciclista = 4
tiempo_encuentro = math.lcm(primer_ciclista, segundo_ciclista)

print(f"Se encontrarán en {tiempo_encuentro} minutos o {round(tiempo_encuentro/60,2)} horas")
print(f"El primer ciclista habrá dado {int(round(tiempo_encuentro / primer_ciclista ))} vueltas")
print(f"El segundo ciclista habrá dado {int(round(tiempo_encuentro / segundo_ciclista))} vueltas")
```

```python
# 9) Fabián tiene un reloj que da una señal cada 18 minutos, otro que da una señal cada 12 minutos y un tercero cada 42 minutos. A las 11 de la mañana los 3 relojes han coincidido en dar la señal, ¿cuántos minutos como mínimo han de pasar para que vuelvan a coincidir?, ¿a qué hora volverán a dar la señal otra vez juntos?

reloj_uno = 18
reloj_dos = 12
reloj_tres = 42

coincidencia = math.lcm(reloj_uno, reloj_dos, reloj_tres)
coincidencia_hrs = coincidencia / 60

print(f"Van a pasar {coincidencia} minutos u {coincidencia_hrs} horas para que vuelvan a coincidir")
print(f"Van a volver a coincidir a las {int(11 + coincidencia_hrs)} horas y {int(60 * .2)} minutos")
```
