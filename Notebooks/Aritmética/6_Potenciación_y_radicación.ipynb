## **Suma y Resta**

```python
x = 2 * sp.root(5,3) + 11 * sp.root(5,3)
resultado = sp.simplify(x)
print(resultado)
```

```python
x = 3 * sp.root(2,2) + 7 * sp.root(2,2) - 4 * sp.root(2,2)
resultado =  sp.simplify(x)
print(resultado)
```

```python
x = (sp.Rational(3,4) * sp.sqrt(6)) - sp.Rational(1,6) * sp.sqrt(6)
resultado = sp.simplify(x)
print(resultado)
```

```python
x = sp.sqrt(20) + sp.sqrt(45) - sp.sqrt(80)
resultado = sp.simplify(x)
print(x)
```

```python
x = sp.root(189, 3) + sp.root(56,3)
resultado = sp.simplify(x)
print(x)
```
